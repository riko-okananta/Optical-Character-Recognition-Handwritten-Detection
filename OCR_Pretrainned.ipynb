```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import requests
from google.colab import files
import matplotlib.pyplot as plt
```

## Model 1

Penjelasan Model :
1. Ukuran model 2.23gb
2. file input(ukuran gambar) yang dapat masuk ke model (248,76)

```python
processor = TrOCRProcessor.from_pretrained('microsoft/trocr-large-handwritten')
model = VisionEncoderDecoderModel.from_pretrained('microsoft/trocr-large-handwritten')
# pixel_values = processor(images=image, return_tensors="pt").pixel_values

# generated_ids = model.generate(pixel_values)
# generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

### Test Model 1

```python
# Mengunggah gambar
uploaded = files.upload()

# Ambil gambar yang diunggah
uploaded_image = list(uploaded.keys())[0]
# Load the image file
img = Image.open(uploaded_image).convert('RGB')  # Convert to grayscale

img = img.resize((248, 76))

pixel_values = processor(images=img, return_tensors="pt").pixel_values

generated_ids = model.generate(pixel_values)
generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

# Menampilkan gambar
plt.imshow(img)
plt.axis('off')
plt.show()

print("Prediksi tulisan:", generated_text)
```
